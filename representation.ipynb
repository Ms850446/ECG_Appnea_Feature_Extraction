```python
import wfdb
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.fft import fft, fftfreq
```

```python

appnea=[]
data=[]
ecg=[['a','b','c'],[20,5,10]]
for iterate in range(0,3):
    x='0'
    size=ecg[1][iterate]
    for i in range(1,size+1):
        if(i==10):
            x=''
        signal= wfdb.rdrecord(record_name=("apnea-ecg-database-1.0.0/"+ecg[0][iterate]+ x + str(i) ) ).p_signal
        signal = signal.reshape(len(signal))
        annotation = wfdb.rdann(("apnea-ecg-database-1.0.0/"+ecg[0][iterate]+ x + str(i) ), extension="apn")
        recordSample=int(len(signal)/len(annotation.symbol))
        for j in range(0,recordSample*len(annotation.symbol), recordSample):
            data.append(signal[j:j+recordSample])
        for element in annotation.symbol:
            appnea.append(element)

# appnea=np.array(appnea).ravel()
print(len(appnea))
print(len(data))



```

```python
FEATURES = ['MIN','MAX','MEAN','RMS','VAR','STD','POWER','PEAK','P2P','CREST FACTOR','SKEW','KURTOSIS',
            'MAX_f','SUM_f','MEAN_f','VAR_f','PEAK_f','SKEW_f','KURTOSIS_f',"Appnea"]

Min=[];Max=[];Mean=[];Rms=[];Var=[];Std=[];Power=[];Peak=[];Skew=[];Kurtosis=[];P2p=[];CrestFactor=[];
FormFactor=[]; PulseIndicator=[];
Max_f=[];Sum_f=[];Mean_f=[];Var_f=[];Peak_f=[];Skew_f=[];Kurtosis_f=[]
```

```python
print(len(data))
i=0
for signal in data:
    print(len(signal))
    i=i+1
    X = signal
    ## TIME DOMAIN ##

    Min.append(np.min(X))
    Max.append(np.max(X))
    Mean.append(np.mean(X))
    Rms.append(np.sqrt(np.mean(X**2)))
    Var.append(np.var(X))
    Std.append(np.std(X))
    Power.append(np.mean(X**2))
    Peak.append(np.max(np.abs(X)))
    P2p.append(np.ptp(X))
    CrestFactor.append(np.max(np.abs(X))/np.sqrt(np.mean(X**2)))
    Skew.append(stats.skew(X))
    Kurtosis.append(stats.kurtosis(X))
    FormFactor.append(np.sqrt(np.mean(X**2))/np.mean(X))
    PulseIndicator.append(np.max(np.abs(X))/np.mean(X))
    ## FREQ DOMAIN ##
    ft = fft(X)
    S = np.abs(ft**2)/len(X)
    Max_f.append(np.max(S))
    Sum_f.append(np.sum(S))
    Mean_f.append(np.mean(S))
    Var_f.append(np.var(S))
    
    Peak_f.append(np.max(np.abs(S)))
    Skew_f.append(stats.skew(X))
    Kurtosis_f.append(stats.kurtosis(X))
```

```python
df_features = pd.DataFrame(index = [FEATURES], 
                               data = [Min,Max,Mean,Rms,Var,Std,Power,Peak,P2p,CrestFactor,Skew,Kurtosis,
                                       Max_f,Sum_f,Mean_f,Var_f,Peak_f,Skew_f,Kurtosis_f,appnea])
df_features=df_features.transpose()
df_features
```

```python
df_features.to_csv("isAppneaFeatures.csv")
```

```python
import wfdb
import pandas as pd

# set the path to your data files
recordname = 'apnea-ecg-database-1.0.0/a01'
ann_name = 'apnea-ecg-database-1.0.0/a01'

# read the data and annotations
record = wfdb.rdsamp(recordname)
annotation = wfdb.rdann(ann_name, extension='apn')

# resample the data to 6000 samples
record_resampled, _ = wfdb.processing.resample_sig(record.p_signals, record.fs, fs=6000)

# get the annotations corresponding to the resampled data
ann_resampled = wfdb.Annotation(record_name='', extension='apn', sample=annotation.sample//record.fs*6000, 
                                 symbol=annotation.symbol, subtype=annotation.subtype, aux_note=annotation.aux_note)

# create a dataframe from the resampled data and annotations
df = pd.DataFrame(data=record_resampled, columns=record.sig_name)
df['annotation'] = ann_resampled.symbol

# save the dataframe as a csv file
df.to_csv('output.csv', index=False)
```
